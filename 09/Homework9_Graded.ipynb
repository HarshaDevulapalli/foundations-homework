```python
def depth_to_words(depth_in_numbers):
    
    if int(depth_in_numbers) < 70:
        return "Shallow"
    elif int(depth_in_numbers) > 70 and depth_in_numbers<300:
        return "Intermediate"
    elif int(depth_in_numbers)>300:
        return "Deep"

```

```python
def power_to_words(power_in_numbers):
    
    if float(power_in_numbers) < 2.5:
        return "Unnoticable Tremors were"
        
    elif float(power_in_numbers) > 2.5 and float(power_in_numbers) < 5.5:
        return "Mild Tremors were"
        
    elif float(power_in_numbers) >= 5.5 and float(power_in_numbers) < 6.0:
        return "Minor Earthquake was"
        
    elif float(power_in_numbers) >= 6.0 and float(power_in_numbers) < 6.9:
        return "Earthquake was"
        
    elif float(power_in_numbers) >= 6.9 and float(power_in_numbers) < 7.9:
        return " Major Earthquake was"
        
    elif float(power_in_numbers)>=7.9:
        return " Massive Earthquake was"
        
```

```python
def date_and_words(time):
    import dateutil.parser
    date = dateutil.parser.parse(time)
    month_list=['January','February','March','April','May','June','July','August','September','October','November','December']
    month_name= month_list[(date.month)-1]
    print(date.day,month_name)
    
```

```python
def time_of_the_day(time):
    import dateutil.parser
    date=dateutil.parser.parse(time)
    if (date.hour) >= 6 and (date.hour) <= 12:
        timeofday="morning" 
    elif (date.hour) > 12 and (date.hour) <= 18:
        timeofday="afternoon"
    elif (date.hour) >18 and (date.hour) <=22:
        timeofday="evening"
    else:
        timeofday="night"
    return timeofday
```

```python
def day_to_words(time):
    import dateutil.parser
    date=dateutil.parser.parse(time)
    return date.strftime("%A")
```
